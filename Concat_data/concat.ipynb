```python
def move_files(source_dir, new_dir, data):
    files = os.listdir(source_dir)
    for file in files:
        # Construct the source file path
        source_file = os.path.join(source_dir, file)

        # Generate the new file name
        new_name = data + file

        # Construct the destination file path
        destination_file = os.path.join(new_dir, new_name)

        # Rename and move the file to the new directory
        shutil.copy(source_file, destination_file)

        print(f"File '{file}' renamed and saved as '{new_name}' in '{new_dir}'.")
```

```python
def energy_identifier(structure_dir, source):
    files = os.listdir(structure_dir)
    idx = []
    for file in files:
        #print(file)
        idx.append(file)
    numbers = [int(re.search(r'\d+', string).group()) for string in idx if re.search(r'\d+', string)]
    numbers = sorted(numbers)
    files = [source+f'_{i}.vasp' for i in numbers]
    return files
```
